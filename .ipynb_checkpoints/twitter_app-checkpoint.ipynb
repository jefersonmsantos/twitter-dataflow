```python
#!py -m pip install tweepy
import tweepy
from datetime import datetime
import json
import pandas as pd
```

```python
#Realizar autenticação no twitter
auth = tweepy.OAuthHandler(API_Key, API_Secret_key)
auth.set_access_token(access_token, access_token_secret)

#construir instância api
api = tweepy.API(auth, wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python
api.update_status('HI test2')
```

```python
#Busca por palavras chave
#keyword = "'disney plus' OR 'disney+' OR 'disney +' OR '#disneyplus' OR '#disney+' OR 'netflix' OR '#netflix' OR 'globoplay' OR '#GLOBOPLAY' OR 'hbogo' OR 'hbo go' OR '#HBOGO' OR 'amazon prime' OR 'AMAZON PRIME VIDEO' OR '#primevideo' OR 'amazonprime'"
keyword = "'NETFLIX' OR 'HBOGO' OR 'DISNEYPLUS' OR 'DISNEY+' OR 'GLOBOPLAY' OR '#GLOBOPLAY'"
brasil_geo_code = api.geo_search(query="Brazil", granularity="country")[0].id

data_agora = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
out = open(f"collected_tweets_{data_agora}.txt","w")

for tweet in tweepy.Cursor(api.search,
                          q='{} place:{}'.format(keyword, brasil_geo_code), 
                          #q=keyword,  
                          tweet_mode='extended',
                          rpp=2000,
                          result_type="recent",
                          lang='pt',
                          include_entities=True,
                          include_rts=False).items(10):
    
    itemstring=json.dumps(tweet._json)
    out.write(itemstring + '\n')
```
